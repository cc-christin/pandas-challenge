```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head()
# purchase_data, 779 rows
```

## Player Count

* Total Number of Players in Heroes of Pymoli.




```python
players_df = purchase_data.loc[:, ["SN", "Age", "Gender"]]

# remove duplicates
players_df = players_df.drop_duplicates()
total_players = players_df.drop_duplicates()
# total_players = players_df
# keeping count of players
total_players = total_players.count()[0]

# display total player count
print(f'Total Heroes Of Pymoli Players: {total_players}')
```

## Purchasing Analysis (Total)

* Number of Unique Items
* Average Purchase Price
* Total Number of Purchases
* Total Revenue 


```python
# Basic Calculations
# number of unique items
unique_items = purchase_data["Item ID"].nunique()
# average price
average_price = purchase_data["Price"].mean()
# number of purchases
number_purchases = len(purchase_data)
# total revenue
total_revenue = purchase_data["Price"].sum()

# DataFrame: purchase_analysis_df
purchase_analysis_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                                    "Average Price": [average_price],
                                    "Number of Purchases": [number_purchases],
                                    "Total Revenue": [total_revenue]})
# Formatting DataFrame, dollar signs
pd.options.display.float_format = '${:,.2f}'.format

# Display Summary DataFrame
purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Calculations of Gender Demographics: Percentage and Count of Male, Female, and Other/ Non-Disclosed Players

# Duplicates have already been dropped from players_df
# Total Gender Demographics 
total_demo = players_df["Gender"].value_counts()

# Demographic Percentages
percentage_demo = total_demo / total_players * 100

# Gender Demograpics DataFrame
gender_df = pd.DataFrame({"Percentage of Players": percentage_demo,
                          "Total Count": total_demo})

# Format DataFrame, percentages
pd.options.display.float_format = '{:,.2f}%'.format

# Display DataFrame
gender_df
```


## Purchasing Analysis (Gender)

#### Basic Calculations For Purchasing Analysis By Gender
* Purchase Count
* Average Purchase Price 
* Total Purchase Value 
* Avg Total Purchase per Person

#### DataFrame
* gender_purchase_analysis_df

#### Formating 
* pd.options.display.float_format = '{:,.2f}'.format
* pd.options.display.float_format = '${:,.2f}'.format

```python
# Grouping Data by Gender Using .groupby()
gender_grouping_df = purchase_data.groupby(["Gender"])

# Calculating Purchase Count Using .count()
gender_grouping_df["Purchase ID"].count().head(10)

# Calculating Total Purchase Value By Gender Using .sum()
total_purchase_value = gender_grouping_df["Price"].sum()
total_purchase_value.head(10)
# Formatting total_purchase_value
format_total_purchase_value = total_purchase_value.map("${:,.2f}".format)
format_total_purchase_value.head(10)

# Calculating Average Purchase Price By Gender using .mean()
average_purchase_price = gender_grouping_df["Price"].mean()
average_purchase_price.head(10)
# Formatting average_purchase_price
format_average_purchase_price = average_purchase_price.map("${:,.2f}".format)
format_average_purchase_price.head(10)

# Calculating Normalized (Avg Total Purchase per Person)
avg_total_per = total_purchase_value / gender_grouping_df["Purchase ID"].count()
# Formating Normalized Totals
format_avg_total_per = avg_total_per.map("${:,.2f}".format)
format_avg_total_per.head(10)

# Ordered DataFrame For Purchasing Analysis Calculations By Gender
ordered_gender_purchase_df = pd.DataFrame(gender_grouping_df["Purchase ID"].count())
ordered_gender_purchase_df["Average Purchase Price"] = format_average_purchase_price 
ordered_gender_purchase_df["Total Purchase Value"] = format_total_purchase_value
ordered_gender_purchase_df["Avg Total Purchase per Person"] = format_avg_total_per

# Display Summary DataFrame For Purchasing Analysis Grouped By Gender
gender_purchase_analysis_df = ordered_gender_purchase_df.rename(columns={"Purchase ID":"Purchase Count"})
gender_purchase_analysis_df
```
